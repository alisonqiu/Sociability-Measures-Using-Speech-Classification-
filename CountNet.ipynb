```python
import numpy as np
import soundfile as sf
import argparse
import os
#import keras
import sklearn
import librosa

def load_audio_files(path: str):

        dataset = []
        walker = sorted(str(p) for p in Path(path).glob(f'*.wav'))
        for i, file_path in enumerate(walker):
                #seperate file_path of each recording into yes folder's path and each recording's name  
                path, filename = os.path.split(file_path)
                label = int(filename.split("_")[0])
                print(label)

                audio, rate = sf.read(file_path, always_2d=True)
                audio = np.mean(audio, axis=1)
                #print(audio.shape)
                # load preprocessing parameters
                scaler = sklearn.preprocessing.StandardScaler()
                # save scaler.npz from https://github.com/faroit/CountNet/tree/master/models in e.g., /content/scalaer.npz
                with np.load('/Users/alisonqiu/Downloads/LibriCount10-0dB/test2/scaler.npz') as data:
                        #print(data)
                        scaler.mean_ = data['arr_0']
                        scaler.scale_ = data['arr_1']

                # compute STFT
                X = np.abs(librosa.stft(audio, n_fft=400, hop_length=160)).T

                # apply global (featurewise) standardization to mean1, var0
                X = scaler.transform(X)

                # cut to input shape length (500 frames x 201 STFT bins)
                X = X[:500, :]

                # apply l2 normalization
                eps = np.finfo(float).eps
                Theta = np.linalg.norm(X, axis=1) + eps
                X /= np.mean(Theta)

                # add sample dimension
                X = X[np.newaxis, ...]

                # if your input requires 4d input:
                #X = X[:, np.newaxis, ...]

                        #[tensor, 16000, '0_006e22.wav'] 
                dataset.append([X, rate, label]) 

        return dataset
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Using {} device'.format(device))

class CountNet(nn.Module):
    def __init__(self):
        super(CountNet,self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1,out_channels=64,kernel_size=3)
        self.conv2 = nn.Conv2d(in_channels=64,out_channels=32,kernel_size=3)
        self.maxpool = nn.MaxPool2d(3)
        self.conv3 = nn.Conv2d(in_channels=32,out_channels=128,kernel_size=3)
        self.conv4 = nn.Conv2d(in_channels=128,out_channels=64,kernel_size=3)
        self.dropout = nn.Dropout(p=0.25)
        #change input size
        self.lstm = nn.LSTM(input_size=64*20, hidden_size=40, batch_first=True)
        self.maxpool2 = nn.MaxPool1d(2)
        self.flatten = nn.Flatten()
        self.fc = nn.Linear(1040 , 3)

    def forward(self, x):
        x = self.conv1(x)
        #print("after conv1",x.shape)
        x = self.conv2(x)
        #print("after conv2",x.shape)
        x = self.maxpool(x)
        #print("after maxpool",x.shape)
        x = self.conv3(x)
        #print("after conv3",x.shape)
        x = self.conv4(x)
        #print("after conv4",x.shape)
        x = self.maxpool(x)
        #print("after maxpool",x.shape)
        x = self.dropout(x)
        #print("after dropout",x.shape) #[1, 64, 53, 20]
        x = x.permute(0, 2, 1,3) #[1, 53, 64, 20]
        x = torch.reshape(x, (x.size(0), x.size(1),-1)) #[1,53,64*20]
        #print("after reshape",x.shape) #[1, 53, 1280]

        x,_ = self.lstm(x)
        #print("after lstm",x.shape)#[1, 53, 40]
        x = x.permute(0, 2, 1)
        x = self.maxpool2(x)
        x = x.permute(0, 2, 1)
        #print("after maxpool2",x.shape) #[1, 53, 20]
        x = self.flatten(x)
        #print("after flatten",x.shape)
        #x = torch.reshape(x, (x.size(0), -1))
        #x= x[:,-1,:]
        x= self.fc(x)
        #print(x,"final shape",x.shape)
        return x 

        

model = CountNet().to(device)
```

```python

batch_size = 1
summary(model, input_size=(batch_size, 1, 500, 201))
```

```python
# cost function used to determine best parameters
cost = torch.nn.CrossEntropyLoss()

# used to create optimal parameters
learning_rate = 0.0001

optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
#scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=1, gamma=0.1)

# Create the training function

def train(dataloader, model, loss, optimizer):
    model.train()
    size = len(dataloader.dataset)
    for batch, (X, Y,Z) in enumerate(dataloader):
        
        X, Z = X.to(device), Z.to(device)
        optimizer.zero_grad()
        pred = model(X.float())
        print(pred.argmax(1))
        loss = cost(pred, Z)
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * len(X)



# Create the validation/test function

def test(dataloader, model):
    model.eval()
    test_loss, correct = 0, 0
    size = len(dataloader.dataset)
    with torch.no_grad():
        for batch, (X, Y,Z) in enumerate(dataloader):
        
            X, Z = X.to(device), Z.to(device)
            pred = model(X.float())

            test_loss += cost(pred, Z).item()
            correct += (pred.argmax(1)==Z).type(torch.float).sum().item()

    test_loss /= size
    correct /= size

    print(f'\nTest Error:\nacc: {(100*correct):>0.1f}%, avg loss: {test_loss:>8f}\n')
```

```python
epochs = 5

for t in range(epochs):
    print(f'Epoch {t+1}\n-------------------------------')
    train(train_dataloader, model, cost, optimizer)
    #scheduler.step()
    test(test_dataloader, model)
print('Done!')
```
